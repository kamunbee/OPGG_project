```python
import pandas as pd
import re
```

### LCK 경기 데이터중 담원이 포지션을 이동해 경기를 했던 적이있어서
### 데이터가 중복으로 들어가는 것을 확인

```python
lck = pd.read_csv('projectData/LCK1.csv', encoding='cp949', index_col=0)
lck_po = pd.read_csv('projectData/LCK_PO.csv', index_col=0)


# # 쇼메이커, 캐니언 겹치는 정보 드랍해주기
lck1 = lck[lck['call_name'] == 'DK ShowMaker']
lck2 = lck[lck['call_name'] == 'DK Canyon']
a =lck1[lck1['line'] == 'AD'].index
b =lck2[lck2['line'] == 'MID'].index

lck = lck.drop(index=a)
lck = lck.drop(index=b)
```

### 중복된 데이터의 인덱스값을 드랍해주는 형태로 제거

```python
# PO 데이터까지 포함하기위해 데이터를 합침
LCK =pd.concat([lck,lck_po])
```

## 다른 리그 데이터 모두 가져오기
### LPL과 VCS는 데이터가 없어서 제외

```python
CBLOL = pd.read_csv('projectData/CBLOL1.csv',index_col='Unnamed: 0')
LCL = pd.read_csv('projectData/LCL1.csv',index_col='Unnamed: 0')
LCO = pd.read_csv('projectData/LCO1.csv',index_col='Unnamed: 0')
LCS = pd.read_csv('projectData/LCS1.csv',index_col='Unnamed: 0')
LEC = pd.read_csv('projectData/LEC1.csv',index_col='Unnamed: 0')
LJL = pd.read_csv('projectData/LJL1.csv',index_col='Unnamed: 0')
LLA = pd.read_csv('projectData/LLA1.csv',index_col='Unnamed: 0')
TCL = pd.read_csv('projectData/TCL1.csv',index_col='Unnamed: 0')
PCS = pd.read_csv('projectData/PCS1.csv',index_col='Unnamed: 0')
```

### 모든 리그 데이터를 합침

```python
# 리그 데이터 합쳐주기
All_league = pd.concat([LCK,CBLOL,LCL,LCO,LCS,LEC,LJL,LLA,TCL,PCS])
```

```python
# 중복값을 제거해 하나의 선수 이름값만 남김
All_league_player = All_league['player_name'].unique()
All_league_player
```

```python
# 데이터를 담을 빈 리스트 작성
korean_player = []

# 반복문으로 리스트안에 각 데이터를 넣어줌
for i in All_league_player:
    korean_player.append(i)
```

```python
# 한국어만 남기도록 처리해줌
korea = re.compile('[^ ㄱ-ㅣ가-힣]+')

# 한국어와 스플릿으로 구분된 빈칸 제거
result = korea.sub('',str(korean_player))
result2 = result.replace(' ','')
result2 = result2.replace("'",'')

# 한국인은 일반적으로 이름이 3글자니까 str 전체 길이에서 3으로 나누어줌
len(result2)/3

## 122.66666 인 이유는 허수라는 두글자이름이 있기때문에
```
