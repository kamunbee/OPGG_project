```python
# selenium을 백그라운드로 실행하기 위해 옵션 설정
options = webdriver.ChromeOptions()
# options.add_argument('headless')
options.add_argument('--start-fullscreen')
# options.add_argument('window-size=1920x1080')
```

```python
# 우리가 데이터를 가져올 웹 페이지입니다.
qwergg_url = 'https://lolalytics.com/lol/tierlist/?patch=11.19'

# selenium이 제어할 chrome을 실행합니다.
driver = webdriver.Chrome("C:/Users/LUNA/Desktop/OPGG/chromedriver_win32/chromedriver")
# driver = webdriver.Chrome("C:/Users/LUNA/Desktop/OPGG/chromedriver_win32/chromedriver", options= options)
```

```python
# 데이터를 가져올 페이지로 이동합니다.
driver.get(qwergg_url)
```

```python
rank = []
champion = []
tier = []
win_rate=[]
pick = []
ban  = []
pbi  = []
games = []
version=[]
```

# lol analyst 사이트에서 글로벌 솔랭 지표 가져오기

```python
time.sleep(5)
a = driver.find_elements_by_css_selector('#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div')


for i in range(2,len(a)+1):
    
    # 버전 값 넣어주기
    version.append('11.19')
    # 글로벌 지표들 가져오기
    rank.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(1)').text.strip())
    champion.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(3) ').text.strip())
    tier.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(4) ').text.strip())
    win_rate.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(6) > div:nth-child(1)').text.strip())
    pick.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(7)').text.strip())
    ban.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(8)').text.strip())
    pbi.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(9)').text.strip())
    games.append(driver.find_element_by_css_selector(f'#root > div > div.Wrapper_content__1wGv- > div:nth-child(5) > div > div:nth-child({i}) > div:nth-child(10)').text.strip())
```

```python
total_df = pd.DataFrame([version,
                        rank,
                        champion,
                        tier,
                        win_rate,
                        pick,
                        ban,
                        pbi,
                        games],
                        
                        index = ['version',
                        'rank',
                        'champion',
                        'tier',
                        'win_rate',
                        'pick',
                        'ban',
                        'pbi',
                        'games']).T
```

## 11.11부터 11.16,  11.19까지 반복
